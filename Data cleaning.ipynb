```python
#Import dataset
elections = pd.read_csv('2020 November General.csv')
elections.head()
```

```python
#review office titles to find relevant office
elections['OfficeTitle'].value_counts()
```

```python
#Filter for preferred office
elections_p = elections[elections['OfficeTitle'] == 'President and Vice President']
```

```python
#loop over localities for sum of votes for each locality
for LocalityName in elections_p:
    total_votes = elections_p.groupby('LocalityName').sum('TOTAL_VOTES')

print(total_votes)
```

```python
#Loop over localities for sum of votes by party in each locality
for LocalityName in elections_p:
    party_votes = elections_p.groupby(['LocalityName', 'Party']).sum('TOTAL_VOTES').reset_index()
print(party_votes)   
```

```python
#create new dataframe 
Votes = pd.DataFrame(columns=['Total', 'Dem_votes', 'Rep_votes', 'Lib_votes', 'Dem_votes%', 'Rep_votes%', 'Lib_votes%'])
```

```python
#populate vote totals column
Votes['Total'] = total_votes['TOTAL_VOTES']
```

```python
#review to ensure all parties have same number of entries
party_votes['Party'].value_counts()
```

```python
#filter for party, merge with new dataframe, populate party column, drop extra column
democrat_votes = party_votes[['LocalityName', 'TOTAL_VOTES']].where(party_votes['Party'] == 'Democratic').dropna()
Votes = Votes.merge(democrat_votes, how='left', on='LocalityName')
Votes['Dem_votes'] = Votes['TOTAL_VOTES']
Votes.drop(['TOTAL_VOTES'], axis=1, inplace=True)
```

```python
republican_votes = party_votes[['LocalityName', 'TOTAL_VOTES']].where(party_votes['Party'] == 'Republican').dropna()
Libertarian_votes = party_votes[['LocalityName', 'TOTAL_VOTES']].where(party_votes['Party'] == 'Libertarian').dropna()
```

```python
Votes = Votes.merge(republican_votes, how='left', on='LocalityName')
Votes['Rep_votes'] = Votes['TOTAL_VOTES']
Votes.drop(['TOTAL_VOTES'], axis=1, inplace=True)
```

```python
Votes = Votes.merge(Libertarian_votes, how='left', on='LocalityName')
Votes['Lib_votes'] = Votes['TOTAL_VOTES']
Votes.drop(['TOTAL_VOTES'], axis=1, inplace=True)
Votes
```

```python
#replace all NaNs with 0
Votes['Dem_votes'].replace('NaN', 0, inplace=True)
Votes['Rep_votes'].replace('NaN', 0, inplace=True)
Votes['Lib_votes'].replace('NaN', 0, inplace=True)
```

```python
#calculate vote percentages by party
Votes['Dem_votes%'] = Votes['Dem_votes']/Votes['Total'].round(2)*100
Votes['Rep_votes%'] = Votes['Rep_votes']/Votes['Total'].round(2)*100
Votes['Lib_votes%'] = Votes['Lib_votes']/Votes['Total'].round(2)*100
```

```python
#re-order counties and cities with same name to have county first to match other datasets
fxci_copy = Votes.iloc[37].copy()
fxco_copy = Votes.iloc[38].copy()
Votes.iloc[37] = fxco_copy
Votes.iloc[38] = fxci_copy

frci_copy = Votes.iloc[43].copy()
frco_copy = Votes.iloc[44].copy()
Votes.iloc[43] = frco_copy
Votes.iloc[44] = frci_copy

rici_copy = Votes.iloc[104].copy()
rico_copy = Votes.iloc[105].copy()
Votes.iloc[104] = rico_copy
Votes.iloc[105] = rici_copy

roci_copy = Votes.iloc[106].copy()
roco_copy = Votes.iloc[107].copy()
Votes.iloc[106] = roco_copy
Votes.iloc[107] = roci_copy
```

```python
Votes
```

```python
#Save cleaned dataset
Votes.to_csv('votes_wk8.csv')
```

```python
#Import dataset
cases = pd.read_csv('VDH-COVID-19-PublicUseDataset-Cases-4.csv')
cases.head(10)
```

```python
#Group by locality and choose max for each locality
cases = cases.groupby('Locality').max('Total Cases')
cases = cases.reset_index()
cases.head(10)
```

```python
len(cases)
```

```python
#re-order counties and cities with same name to have county first to match other datasets
frci_copy = cases.iloc[43].copy()
frco_copy = cases.iloc[44].copy()
cases.iloc[43] = frco_copy
cases.iloc[44] = frci_copy

rici_copy = cases.iloc[104].copy()
rico_copy = cases.iloc[105].copy()
cases.iloc[104] = rico_copy
cases.iloc[105] = rici_copy

roci_copy = cases.iloc[106].copy()
roco_copy = cases.iloc[107].copy()
cases.iloc[106] = roco_copy
cases.iloc[107] = roci_copy
```

```python
#Save cleaned dataset
cases.to_csv('cases_wk8.csv')
```
