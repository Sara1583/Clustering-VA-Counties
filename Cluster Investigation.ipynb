```python
#import libraries and set style
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
sns.set_style('darkgrid')
```

```python
#import final dataset, set county as index
clusters = pd.read_csv('data_with_clusters_wk8.csv')
clusters = clusters.set_index('County', drop=True)
clusters.head()
```

```python
#create new dataframe of averages of each characteristics within each cluster
cluster_avg = clusters.groupby('Cluster').mean()
cluster_avg = pd.DataFrame(cluster_avg)
cluster_avg = cluster_avg.reset_index()
```

```python
cluster_avg
```

Create barplots for select characteristics and identify potential target clusters

```python
sns.barplot(data=cluster_avg, y='%_65+', x='Cluster')
plt.title('Average of population over 65 by Cluster')
plt.savefig('Average of population over 65 by Cluster')
plt.show()
print('Target clusters: 2, 5, 9','\n')
print('Cluster 2:', '\n', np.array(clusters[clusters['Cluster'] == 2].index),'\n')

print('Cluster 5:', '\n', np.array(clusters[clusters['Cluster'] == 5].index),'\n')

print('Cluster 9:', '\n', np.array(clusters[clusters['Cluster'] == 9].index))
```

```python
sns.barplot(data=cluster_avg, y='%_Black', x='Cluster')
plt.title('Average of population that is Black by Cluster')
plt.savefig('Average of population that is Black by Cluster')
plt.show()
print('Target clusters: 3, 4, 9','\n')
print('Cluster 3:', '\n', np.array(clusters[clusters['Cluster'] == 3].index),'\n')

print('Cluster 4:', '\n', np.array(clusters[clusters['Cluster'] == 4].index),'\n')

print('Cluster 9:', '\n', np.array(clusters[clusters['Cluster'] == 9].index))
```
