```python
#Import libraries and each cleaned dataset

import pandas as pd
#pd.set_option('display.max_columns', 140)
#pd.set_option('display.max_rows', 140)
```

```python
cases = pd.read_csv('cases_wk8.csv')
print(len(cases))
cases.head(10)
```

```python
demographics = pd.read_csv('demographics_wk8.csv')
print(len(demographics))
demographics.head(10)
```

```python
poverty = pd.read_csv('Poverty_est_wk8.csv')
print(len(poverty))
poverty.head(10)
```

```python
votes = pd.read_csv('votes_wk8.csv')
print(len(votes))
votes.head(10)
```

```python
#create dataframe with specific desired features
data = pd.DataFrame(columns=['County', '%_65+', '%_Black', '%_Indigenous', '%_Asian', '%_Pacific_Is', 
                             '%_Hispanic', '%_White', '%_Rural', '%_Est_poverty', '%_Dem_votes', '%_Rep_votes',
                             '%_Lib_votes', 'Cases/Pop'])
```

```python
data
```

```python
#populate dataframe with relevant columns from cleaned datasets
data['County'] = poverty['County']
data[['%_65+']] = demographics[['% 65 and over']]
data['%_Black'] = demographics['% Black']
data['%_Indigenous'] = demographics['% American Indian & Alaska Native']
data['%_Asian'] = demographics['% Asian']
data['%_Pacific_Is'] = demographics['% Native Hawaiian/Other Pacific Islander']
data['%_Hispanic'] = demographics['% Hispanic']
data['%_White'] = demographics['% Non-Hispanic White']
data['%_Rural'] = demographics['% Rural']
data['%_Est_poverty'] = poverty['Est%_poverty_all']
data['%_Dem_votes'] = votes['Dem_votes%']
data['%_Rep_votes'] = votes['Rep_votes%']
data['%_Lib_votes'] = votes['Lib_votes%']
data['Cases/Pop'] = (cases['Total Cases']/demographics['Population'])*100
```

```python
data
```

```python
#Set county as the index
data = data.set_index('County', drop=True)
```

```python
#Fill NaNs with 0
data.fillna(0, inplace=True)
data
```

```python
#Save collated dataset
data.to_csv('data_wk8.csv')
```
